```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = df1['Region'].value_counts()
x
```

```python
plt.pie(x, labels=x.index, autopct='%1.1f%%')
plt.title('Region and its value counts')
plt.show()
```

```python
df1['Area'].value_counts()
```

```python
sns.displot(df1['Area'])
plt.show()
```

```python
df1.info()
```

```python
df2 = df1.rename(columns={'Estimated Unemployment Rate (%)': 'estimated unemployed rate', 
                          'Estimated Employed': 'EE','Estimated Labour Participation Rate (%)':'ELPR'})
```

```python
df2 = df1.rename(columns={df.columns[0]:'State',df.columns[3]:'EstimatedUnemployment Rate',df.columns[4]:'Estimated Employed',
                          df.columns[5]:'Estimated Labour Participation Rate'})
df2
```

```python
df2.columns
```

```python
sns.pairplot(df2)
plt.show()
```

```python
sns.scatterplot(x = 'EstimatedUnemployment Rate', y = 'State',data = df2, hue = 'Area')
plt.show()
```

```python

# Assuming you have a DataFrame named 'df'
heat_maps = df2[['EstimatedUnemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']]
heat_maps = heat_maps.corr()

plt.figure(figsize=(12, 7))
sns.set_context('notebook', font_scale=1)
sns.heatmap(heat_maps, annot=True, cmap='winter')

plt.show()
```

```python
df2.columns= ["State","Date","Frequency","EstimatedUnemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area"]

plt.figure(figsize=(14, 18))
plt.title("Unemployment_rate")
sns.histplot(x="EstimatedUnemployment Rate", hue="State", data=df2)
plt.show()
```
