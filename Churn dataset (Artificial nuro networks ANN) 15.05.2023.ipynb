```python
import numpy as np
import pandas as pd
```

```python
import tensorflow as tf
import keras
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df.info()
```

```python
df = df.drop('Surname',axis=1)
```

```python
df['Geography'].unique()
```

```python
df_dummies = pd.get_dummies(df,drop_first=True)
```

```python
df_dummies
```

```python
X = df_dummies.drop(columns='Exited')
Y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(input_dim=13, units=6 , activation='relu',kernel_initializer="uniform"))

model.add(Dense(input_dim=13, units=6 , activation='relu',kernel_initializer="uniform"))

model.add(Dense(units=1, activation = 'sigmoid',  kernel_initializer="uniform"))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train , batch_size= 10, epochs=50)
```

```python

y_pred_test = model.predict(x_test)
y_pred_test = (y_pred_test > 0.5)
```

```python
y_pred_train = model.predict(x_train)
y_pred_train = (y_pred_train > 0.5)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
print("R2 for train_data",accuracy_score(y_train,y_pred_train))
```

```python
def build_cross_classifier():
    classifier = Sequential()
    classifier.add(Dense(input_dim=13, units=6 , activation='relu',kernel_initializer="uniform"))
    classifier.add(Dense(input_dim=13, units=6 , activation='relu',kernel_initializer="uniform"))
    classifier.add(Dense(units=1, activation = 'sigmoid',  kernel_initializer="uniform"))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
    return classifier
    
    
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
classifier = KerasClassifier(build_cross_classifier, batch_size = 10 , epochs = 50)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier ,X,Y, cv=5)
```

```python
print(accuracies)
accuracies.mean()
```

```python

estimator = KerasClassifier(build_cross_classifier())
param_grid = {
    'batch_size':[10,20,32,45,51],
     'epochs':[50,100],
     'optimizer':['adam','rmsprop']}

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator, param_grid, scoring='accuracy', cv=5)
grid_result = grid.fit(x_train, y_train)
```
